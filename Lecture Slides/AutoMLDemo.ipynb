```python
#using sample code from https://automl.github.io/auto-sklearn/master/
#using sample code from https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html
import scipy
print(scipy.__version__)

import autosklearn.classification
import sklearn.datasets
import sklearn.model_selection
import sklearn.metrics
import time
import sklearn.neighbors

iris = sklearn.datasets.load_iris()

X=iris.data
y=iris.target

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.3, random_state=42)

start_time = time.time()
knn = sklearn.neighbors.KNeighborsClassifier()
knn.fit(X_train, y_train)
y_hat_knn = knn.predict(X_test)
accuracy = sklearn.metrics.accuracy_score(y_test, y_hat_knn)
runtime = time.time() - start_time
print("KNN")
print("Accuracy: " + str(accuracy))
print("Runtime: " + str(runtime))

start_time = time.time()
auto = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=300, per_run_time_limit=30)
auto.fit(X_train, y_train)
y_hat_auto = auto.predict(X_test)
accuracy = sklearn.metrics.accuracy_score(y_test, y_hat_auto)
runtime = time.time() - start_time
print("Auto")
print(auto.sprint_statistics())
print("Accuracy: " + str(accuracy))
print("Runtime: " + str(runtime))
```
